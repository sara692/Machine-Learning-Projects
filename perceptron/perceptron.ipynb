```python
x=df.iloc[:100,[0,2]]
y=df.iloc[:100,4]
y
```

```python
x
```

```python
rows,features=x.shape
features
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
X_train = X_train.astype(float)
y_train = y_train.astype(int)  # if target is 0 / 1
```

```python
print(X_train.dtypes)
#print(y_train.dtypes)
```

```python
n_epochs = 1000
alpha = 0.01
w = np.random.rand(features)
b = 0

for epoch in range(n_epochs):
    for indx, row in enumerate(X_train.values):
        z = np.dot(row, w) + b
        y_pred = 1 if z > 0 else 0
        error = y_train.iloc[indx] - y_pred
        w = w + alpha * error * row
        b = b + alpha * error
```

```python
def predict_fun(x):
    z = np.dot(x,w) + b
    if z>0:
        result=1
    else:
        result=0
    return result
  
```

```python
sample = X_test.iloc[10]
print(y_test.iloc[10])
print(predict_fun(sample))
```

```python
plt.scatter(X_train.iloc[:, 0], X_train.iloc[:, 1], c=y_train, cmap='viridis')
x_vals = np.linspace(X_train.iloc[:, 0].min(), X_train.iloc[:, 0].max(), 100)
y_vals = (-w[0] * x_vals - b) / w[1]
plt.plot(x_vals, y_vals, linewidth=2)  # No color specified (as per rules)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter Plot Colored by Target")
plt.colorbar(label="Target")
plt.show()
```

```python
correct=0
for idx,row in enumerate(X_test.values):
    result=predict_fun(row)
    print(f"result is {result}")
    print(f"data is {y_test.iloc[idx]}")
    if result== y_test.iloc[idx]:
        correct +=1
accuracy =correct/len(X_test)






```
